# Simple SeqTrees Example

This notebook uses the same kind of small, already-preprocessed table shown in the SeqTrees documentation. The goal is to show the core API without preprocessing distractions.

```python
from pathlib import Path
import sys

# When running this notebook from the repository checkout, prefer the local source tree.
repo_root = Path.cwd()
if (repo_root / "src" / "seqtrees").exists():
    sys.path.insert(0, str(repo_root / "src"))

from seqtrees import SequentialTreeSynthesizer
```

```python
data = [
    {"age": 24.5, "sex_code": 0, "income_bin": 1, "risk_code": 0},
    {"age": 31.2, "sex_code": 1, "income_bin": 1, "risk_code": 0},
    {"age": 45.8, "sex_code": 0, "income_bin": 2, "risk_code": 1},
    {"age": 52.1, "sex_code": 1, "income_bin": 2, "risk_code": 1},
    {"age": 67.4, "sex_code": 0, "income_bin": 3, "risk_code": 2},
    {"age": 73.0, "sex_code": 1, "income_bin": 3, "risk_code": 2},
]

data
```

```python
model = SequentialTreeSynthesizer(
    variable_order=["age", "sex_code", "income_bin", "risk_code"],
    tree_backend="auto",
    continuous_strategy="interpolate",
    continuous_columns=["age"],
    discrete_columns=["sex_code", "income_bin", "risk_code"],
    min_samples_leaf=1,
    random_state=7,
)

model.fit(data)
synthetic = model.sample(len(data), random_state=11)

model.get_variable_order(), synthetic
```

```python
list(zip(data, synthetic))
```

SeqTrees is sampling from a sequential factorization. The generated rows are not copies of complete original rows, but each column is generated conditionally on the prefix columns in the chosen order.
